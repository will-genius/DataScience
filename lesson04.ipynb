```python
#regresion and classification are supervised learning(where we test models)
#but clustering is unsupervised learning (no model testing)
# we group data into clusters each of this clusters have similarities

import pandas
df = pandas.read_csv("Mall_Customers.csv")
df.head()
```

```python
#below we split data into clusters 
# Here we want to cluster mall customers by age income and spending score
#age will help us know the graphics 
#income will help us understand the shoping capabilities
# spending score will asist in knowing the spending habits 
#hence we select column 2 to 4
# first we remove all empty rows 
df.dropna(inplace=True)
# we then setup the data
array = df.values
x = array[:, 2:5] # age, income, spending score
#no variable y
#next we fit x into the model 
#the model is called K-means
from sklearn.cluster import KMeans
model = KMeans(n_clusters=10,random_state=42)
model.fit(x)
```

```python
#Next, We Find the Clusters that K means Created - 10 Clusters. K Means shows Each Clusters and Its Means/Average For Age, Income, Spending Score. The Average Per Cluster Helps you know that the records in That Group in close to the means shown.
means = model.cluster_centers_
clusters = pandas.DataFrame(means,columns=["age","income","score"])
clusters
```

```python
#we show all records in cluster 5
df["no"] = model.labels_
cluster5 = df[df["no"]==5]
cluster5
```

```python
df[df["no"]==5].to_csv("cluster5.csv")
```

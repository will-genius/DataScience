```python
#visualisation 
#Seaborn is a library that provides additional visualization tools built on top of Matplotlib.


import seaborn as sb
import matplotlib.pyplot as plt

#check available themes
plt.style.available
```

```python
plt.style.use("classic")

```

```python
sb.countplot(x=df['Gender'], palette = 'magma')
plt.title("Distribution of Gender Status")
plt.xlabel('Gender')
plt.ylabel('Number of Members')
```

```python
import numpy 
sb.barplot(x = 'Gender', y = 'LoanAmount', data = df, estimator=numpy.mean, palette = 'magma', hue='Education')
plt.title("Correlation between gender,loanAmount and education")
plt.xlabel('Gender')
plt.ylabel('Loan Amount')
```

```python
sb.swarmplot(x ='Gender', y ='ApplicantIncome', data = df, palette= 'magma',hue ='Self_Employed')
```

```python
sb.scatterplot(x ='LoanAmount', y ='ApplicantIncome', data = df, palette= 'magma', hue = 'Gender')
```

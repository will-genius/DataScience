```python
import pandas 
ratings_df = pandas.read_csv("ratings.csv")
ratings_df.head()
```

```python
titles_df = pandas.read_csv("movietitles.csv")
titles_df
```

```python
df = pandas.merge(ratings_df,titles_df,on = "item_id")
df
```

```python
df.groupby("title")["rating"].mean().sort_values(ascending=False)
```

```python
df.groupby("title")["rating"].count().sort_values(ascending=False)
```

```python
# we create a data frame for the rating count
ratings_count = pandas.DataFrame(df.groupby("title")["rating"].mean())
ratings_count["num_of_rating"] = df.groupby("title")["rating"].count()
ratings_count.head(50)
```

```python
pivot = df.pivot_table(index = "user_Id",columns= "title" , values="rating")
pivot
```

```python
selected_movie = pivot["Batman (1989)"]
selected_movie
```

```python
# lets find which movies were rated similarly
similar = pivot.corrwith(selected_movie)
# we then present the corelating movies in a data frame
similar_df = pandas.DataFrame(similar,columns=["Correlations"])
# we sort the results in descending order and limit them to 20 records
similar_df.sort_values("Correlations",ascending=False).head(20)
```

```python
# we then show the recomended movies 
similar_df = similar_df.join(ratings_count["num_of_rating"])
similar_df[similar_df["num_of_rating"]>100].sort_values("Correlations", ascending=False).head(20)
```
